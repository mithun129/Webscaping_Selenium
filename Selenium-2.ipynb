### Scrape data for first 100 sneakers you find when you visit flipkart.com and search for “sneakers” in the search field.
##### You have to scrape 4 attributes of each sneaker :
1. Brand
2. Product Description
3. Price
4. discount

```python
#lets now import all the required libraries
import requests
from selenium import webdriver

browser = webdriver.Chrome()

browser.get("https://www.flipkart.com/")
from bs4 import BeautifulSoup
```

```python
search_btn = browser.find_element_by_class_name("_3704LK")
search_btn.send_keys("sneaker")
```

```python
search_btn = browser.find_element_by_class_name ("L0Z3Pu")
search_btn.click()
```

```python
url="https://www.flipkart.com/search?q=sneaker&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
#lets open the webpage through our webdriver
browser.get(url)
```

```python
brands=[]
product_description=[]
price=[]
discount=[]
```

```python
brand_tags = browser.find_elements_by_xpath ("//div[@class='_2WkVRV']")
brand_tags
```

```python
for i in brand_tags:
    brand=i.text
    brands.append(brand)
    
```

```python
brands
```

```python
product_tags = browser.find_elements_by_xpath ("//a[@class='IRpwTa']")
product_tags
```

```python
for i in product_tags:
    product=i.text
    product_description.append(product)
```

```python
product_description
```

```python
price_tags = browser.find_elements_by_xpath ("//div[@class='_30jeq3']")
price_tags
```

```python
for i in price_tags:
    prices=i.text
    price.append(prices)
```

```python
price
```

```python
discount_tags = browser.find_elements_by_xpath ("//div[@class='_3Ay6Sb']")
discount_tags
```

```python
for i in discount_tags:
    dis=i.text
    discount.append(dis)
```

```python
discount
```

```python
print(len(brands), len(discount), len(price), len(product_description))
```
