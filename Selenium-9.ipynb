The data should include “Brand” of the shoes , Short Shoe description, price of the shoe as shown in the below image. Please note that applying the filter and scraping the data , everything should be done through code only and there should not be any manual step.

```python
#lets now import all the required libraries
import requests
from selenium import webdriver
browser = webdriver.Chrome()
browser.get(" https://www.myntra.com/")
from bs4 import BeautifulSoup
```

```python
#finding element for shoes search bar 
search_btn = browser.find_element_by_class_name("desktop-searchBar")
search_btn.send_keys("shoes sneaker")
```

```python
#do click using xpath function
search_btn = browser.find_element_by_xpath ("//span[@class='myntraweb-sprite desktop-iconSearch sprites-search']")
search_btn.click()
```

```python
brands=[] # empty list
shoedesc=[] # empty list
price=[] # empty list
```

```python
#specifying the URL of webpage to be scraped
url="https://www.myntra.com/shoes-sneaker?f=Color%3ABlack_36454f&plaEnabled=false&rf=Price%3A6657.0_13105.0_6657.0%20TO%2013105.0"
```

```python
#lets open the webpage through our webdriver
browser.get(url)
```

```python
#do click using xpath function
price_tags = browser.find_elements_by_xpath ("//div[@class='product-price']")
price_tags
```

```python
for i in price_tags:
    prices=i.text
    price.append(prices)
    
```

```python
price
```

```python
brand_tags = browser.find_elements_by_xpath ("//h3[@class='product-brand']")
brand_tags
```

```python
for i in brand_tags:
    brand=i.text
    brands.append(brand)
    
```

```python
brands
```

```python
shortdesc_tags = browser.find_elements_by_xpath ("//h4[@class='product-product']")
shortdesc_tags
```

```python
for i in shortdesc_tags:
    sht=i.text
    shoedesc.append(sht)
    
```

```python
shoedesc
```

```python
print(len(brands), len(price),  len(shoedesc))
```
