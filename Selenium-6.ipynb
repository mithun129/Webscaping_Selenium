```python
#lets now import all the required libraries
import requests
from selenium import webdriver

browser = webdriver.Chrome()

browser.get("https://www.flipkart.com/")
from bs4 import BeautifulSoup
```

```python
search_btn = browser.find_element_by_class_name("_3704LK")
search_btn.send_keys("sunglasses")
```

```python
search_btn = browser.find_element_by_class_name ("L0Z3Pu")
search_btn.click
```

```python
#specifying the URL of webpage to be scraped
url="https://www.flipkart.com/search?q=sunglasses&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
#lets open the webpage through our webdriver
browser.get(url)
```

```python
brands=[]
product_description=[]
price=[]
discount=[]
```

```python
brand_tags = browser.find_elements_by_xpath ("//div[@class='_2WkVRV']")
brand_tags
```

```python
for i in brand_tags:
    brand=i.text
    brands.append(brand)
    
```

```python
brands
```

```python
product_tags = browser.find_elements_by_xpath ("//a[@class='IRpwTa']")
product_tags
```

```python
for i in product_tags:
    product=i.text
    product_description.append(product)
```

```python
product_description
```

```python
price_tags = browser.find_elements_by_xpath ("//div[@class='_30jeq3']")
price_tags
```

```python
for i in price_tags:
    prices=i.text
    price.append(prices)
    
```

```python
price
```

```python

discount_tags = browser.find_elements_by_xpath ("//div[@class='_3Ay6Sb']")
discount_tags
```

```python
for i in discount_tags:
    dis=i.text
    discount.append(dis)
```

```python
discount
```

```python
import pandas as pd
jobs=pd.DataFrame({})
jobs['brands']=brands[0:38]
jobs['discount']=discount[0:38]
jobs['price']=price[0:38]
jobs['product_description']=product_description[0:38]
```

```python
jobs
```
